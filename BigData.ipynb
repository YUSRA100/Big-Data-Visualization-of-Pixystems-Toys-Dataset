### To statistically analyise which City has most and least amount of the finished good the customers have ordered, we will use groupby

## Analysis of Purchase goods with respect to Location, Quantity and Total Amount

```python
plt.figure(figsize=(18,15))
temp = dfcon3.groupby(['Location', 'Description']).sum().sort_values('Total Amount', ascending=False).reset_index()
#labels=['10%', 'no-bonus', '20%']
sns.barplot(y='Description', x='Total Amount', data=temp, hue='Location',ci=None, dodge=True)
```

*From here we can clearly observe the relation between the goods that are being purchased with respect to locations and total amount. 
Doll eyes are highest in purchase where as Yellow paint is the lowest.*

**Now, lets analyse it across Item quantity instead of Total amount and see what are the results**

```python
plt.figure(figsize=(18,15))
temp = dfcon3.groupby(['Location', 'Description']).sum().sort_values('Item Quantity', ascending=False).reset_index()
#labels=['10%', 'no-bonus', '20%']
sns.barplot(y='Description', x='Item Quantity', data=temp, hue='Location',ci=None, dodge=True)
```

*We can see that Doll eye are on top in both scearios but the Video game case has the lowest purchases ecpecially in Location 300 which is Shanghai Plant*
